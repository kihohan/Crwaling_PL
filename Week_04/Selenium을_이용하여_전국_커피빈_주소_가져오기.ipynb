```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import time
```

```python
chromedriver_dir = r'/Users/hankiho/Downloads/chromedriver'
driver = webdriver.Chrome(chromedriver_dir)
driver.get('https://www.coffeebeankorea.com/store/store.asp')
time.sleep(0.1)
```

```python
loca = driver.find_element_by_class_name('region_srh').click() #지역검색
time.sleep(0.1)
```

```python
driver.find_element_by_id('localTitle').click()
sido = driver.find_element_by_id('storeLocal') #시도검색
li = sido.find_elements_by_tag_name('li')
li[2].click() #도시 클릭
time.sleep(0.1)
```

```python
def ct_add(i):
    loca = driver.find_element_by_class_name('region_srh').click() #지역검색
    time.sleep(3)
    driver.find_element_by_id('localTitle').click()
    sido = driver.find_element_by_id('storeLocal') #시도검색
    li = sido.find_elements_by_tag_name('li')
    li[i].click()
    time.sleep(3)
```

```python
for i in range(12):
    ct_add(i)
    source = driver.page_source
    bs_obj = bs(source,'html')
    obj = bs_obj.find('div','store_tab')
    address = obj.find('div','cont')
    list_address = address.find_all('li')

    for i in range(len(list_address)):
        print (list_address[i].find('p','address').text)
```
