```python
import pandas as pd
import numpy as np

import platform
import matplotlib.pyplot as plt

from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import urllib
import time
import re


tmp1 = 'https://search.naver.com/search.naver?where=kin' # 네이버 지식인 Url
html = tmp1 + '&sm=tab_jum&ie=utf8&query={key_word}&kin_start={num}' # 네이버 지식은 페이지수 결정 Url

response = urlopen(html.format(num = 1, key_word = urllib.parse.quote(keyword)))
soup = bs(response, "html.parser")
tmp = soup.find('ul',class_='type01')
obj = tmp.find_all('li')
```

```python
date = []
question = []
question_content = []
answer = []
for i in range(len(obj)):
    date.append(obj[i].find('dl').find('dd', class_ = 'txt_inline').text) #날짜
    question.append(obj[i].find('dl').find('dt', class_ = 'question').text) #질문
    question_content.append(obj[i].find('dl').select('dd')[1].text) #질문내용
    answer.append(obj[i].find('dl').find('dd', class_ = 'answer').text) #답변
```

```python
df_question = []
df_answer = []
for i in range(len(question)):
    parse = re.sub('질문','', question[i])
    df_question.append(parse)
for i in range(len(question)):
    parse = re.sub('답변','', answer[i])
    df_answer.append(parse)
```

```python
df_date = pd.DataFrame(date,columns = ['제목'])
df_df_question = pd.DataFrame(question,columns = ['질문'])
df_question_content = pd.DataFrame(question_content,columns = ['질문내용'])
df_df_answer = pd.DataFrame(answer,columns = ['답변내용'])

df_naver_kin = pd.concat([df_date,
                        df_df_question,
                        df_question_content,
                        df_df_answer],axis = 1)
```

```python
keyword = str(input('키워드: '))
```

```python
import pandas as pd
import numpy as np

import platform
import matplotlib.pyplot as plt

from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import urllib
import time
import re
from tqdm import tqdm_notebook

date = []
question = []
question_content = []
answer = []

tmp1 = 'https://search.naver.com/search.naver?where=kin' # 네이버 지식인 Url
html = tmp1 + '&sm=tab_jum&ie=utf8&query={key_word}&kin_start={num}' # 네이버 지식은 페이지수 결정 Url


for n in tqdm_notebook(range(1,1000,10)):
    response = urlopen(html.format(num = n, key_word = urllib.parse.quote(keyword)))
    soup = bs(response, "html.parser")
    tmp = soup.find('ul',class_='type01')
    obj = tmp.find_all('li')

    for i in range(len(obj)):
        date.append(obj[i].find('dl').find('dd', class_ = 'txt_inline').text) #날짜
        question.append(obj[i].find('dl').find('dt', class_ = 'question').text) #질문
        question_content.append(obj[i].find('dl').select('dd')[1].text) #질문내용
        answer.append(obj[i].find('dl').find('dd', class_ = 'answer').text) #답변

    df_question = []
    for i in range(len(question)):
        parse = re.sub('질문','', question[i])
        df_question.append(parse)
    df_answer = []
    for i in range(len(question)):
        parse = re.sub('답변','', answer[i])
        df_answer.append(parse)
    
    time.sleep(0.5)

    df_date = pd.DataFrame(date,columns = ['날짜'])
    df_df_question = pd.DataFrame(df_question,columns = ['질문'])
    df_question_content = pd.DataFrame(question_content,columns = ['질문내용'])
    df_df_answer = pd.DataFrame(df_answer,columns = ['답변내용'])

    df_naver_kin = pd.concat([df_date,
                              df_df_question,
                              df_question_content,
                              df_df_answer],axis = 1)
```
