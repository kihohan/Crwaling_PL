```python
import requests
from urllib.parse import urlparse
import csv
```

```python
# 네이버에서 데이터를 검색
def get_api_result(keyword, display, start):
    url = "https://openapi.naver.com/v1/search/news?query=" + keyword \
          + "&display=" + str(display) \
          + "&start=" + str(start)
    result = requests.get(urlparse(url).geturl(),
          headers={"X-Naver-Client-Id":"vjtMs7cyRqPt7gG7hUOn",
                   "X-Naver-Client-Secret":"gXYZPhVV1l"})
    return result.json()
```

```python
result = get_api_result('강남', 50, 1)
```
