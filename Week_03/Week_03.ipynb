```python
import pandas as pd
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import urllib
```

```python
keyword = str(input('키워드: '))
```

```python
naverblog = 'https://search.naver.com/search.naver?where=post' # 네이버 블로그 Url
option = '&sm=tab_jum&ie=utf8&query={key_word}&start={num}' # 키워드, 페이지
html = naverblog + option
```

```python
for n in range(1, 1000, 10):
    response = urlopen(html.format(num = n, key_word = urllib.parse.quote(keyword)))
    soup = bs(response, "html.parser")
    tmp = soup.find('ul',id = 'elThumbnailResultArea')
    obj = tmp.find_all('li')
    
    for i in range(len(obj)):
        print (obj[i].find('a',class_='sh_blog_title _sp_each_url _sp_each_title').text) #블로그 타이틀
```

```python
from tqdm import tqdm_notebook
```

```python
date = []
title = []
content = []
#좋은 검색결과를 제공하기 위해 결과는 1,000건까지만 제공하고 있습니다.
for n in tqdm_notebook(range(1, 1000, 10)):
    response = urlopen(html.format(num = n, key_word = urllib.parse.quote(keyword)))
    soup = bs(response, "html.parser")
    tmp = soup.find('ul',id='elThumbnailResultArea')
    obj = tmp.find_all('li')
    
    for i in range(len(obj)):
        date.append(obj[i].find('dd',class_='txt_inline').text ) #블로그 게시 날짜
        title.append(obj[i].find('a',class_='sh_blog_title _sp_each_url _sp_each_title').text) #블로그 타이틀
        content.append(obj[i].find('dd',class_='sh_blog_passage').text) #블로그 내용
        
    df_date = pd.DataFrame(date,columns = ['날짜'])
    df_title = pd.DataFrame(title,columns = ['제목'])
    df_content = pd.DataFrame(content,columns = ['본문요약본'])

    df_naver_blog = pd.concat([df_date,
                              df_title,
                              df_content],axis = 1)
```
